#### Data Cleaning

```python
# Deleting the  rows with the missing value is one of the methods by which the missing values could be handled, but this census dataset is not large enough and this method has a disadvantage whicg is the deletion of the row culd significantly reduce the number of observations which could significantly affect the model.
# Using the statistical method of replacing the missing value with the most frequest value (mode) is another method of handling the missing variables, this also has a downside of posssibly leading to skewedness or imbalance in the dataset within the categories
# The method to be applied would be dependent on the specific case of missing variable.
```

```python
# Grouping data into features and outcome

feature_names = ['age', 'workclass', 'fnlwgt', 'education', 'education-num',
       'marital-status', 'occupation', 'relationship', 'sex', 'capital-gain',
       'capital-loss', 'hours-per-week', 'native-country']

X = df_census1[feature_names]
y = df_census1.income
```

```python
#dividing the given dataset into test and training set
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
from sklearn.metrics import accuracy_score
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2,
stratify=df_census1.income, random_state=0)
```

```python
#Training the testing KNN classifier
from sklearn.neighbors import KNeighborsClassifier
#train
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train, y_train)
#testing
preds = knn.predict(X_test)
# Evaluate accuracy
print(accuracy_score(y_test, preds))
```
